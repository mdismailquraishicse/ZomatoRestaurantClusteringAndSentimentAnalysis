```python
print('# of nan in Reviewer column :',len(reviews[reviews['Reviewer'].isna()])) # print(number of nan)
df = reviews[reviews['Reviewer'].isna()==False] # drop nan entities for Reviewer
top_review = df['Review'].describe()['top'] # Assign top element from review column
df['Review'] = df['Review'].fillna(top_review) # Impute nan with top review
df.drop(['Time','Pictures'], axis=1, inplace=True) # dropping the columns Time and Pictures because we don't need them in this project.
```
